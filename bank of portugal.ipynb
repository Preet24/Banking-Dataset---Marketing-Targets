```python
data=data.dropna()
data=data.drop("y",axis=1)
data=data.drop("poutcome",axis=1)
data=data.drop("pdays",axis=1)
data=data.drop("previous",axis=1)
data=data.drop(['day', 'month','campaign'],axis=1)
data.head()
```

```python
s=pd.get_dummies(data.job)
a=pd.get_dummies(data.housing)
b=pd.get_dummies(data.education)
c=pd.get_dummies(data.contact)
data["Loan"]=data.apply(lambda x:0 if x.loan=="no" else 1,axis=1)
data["Marital"]=data.apply(lambda x:1 if x.marital=="married" else 0,axis=1)
data["Default"]=data.apply(lambda x:0 if x.default=="no" else 1,axis=1)
data=data.drop(["loan","default","marital"],axis=1)
```

```python
data=data.join(s,rsuffix="_Job")
data=data.join(a,rsuffix="_Housing")
data=data.join(b,rsuffix="_Education")
data=data.join(c,rsuffix="_Conatct")
data=data.drop(['job','education',"housing","contact"],axis=1)
```

```python
Y=data.Y
X=data.drop("Y",axis=1)
X.columns
```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,shuffle=True,random_state=42)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import matplotlib.pyplot as plt
torch.cuda.is_available()
```

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(27, 16)
        self.fc2 = nn.Linear(16, 4)
        self.fc3 = nn.Linear(4, 1)
       
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.sigmoid(self.fc3(x))
        return x
model = Net()
device = torch.device('cuda')
model.to(device)
print(model)
```

```python
import numpy as np
x_train=x_train.to_numpy()
y_train=y_train.to_numpy()
x_train = torch.from_numpy(x_train.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.float32))
params = {'batch_size': 64,
          'shuffle': False,
          'num_workers': 6}
training_set=torch.utils.data.TensorDataset(x_train,y_train)
training_generator = torch.utils.data.DataLoader(training_set, **params)
```

```python
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
num_epochs = 101
losses=[]
Loss_y=[]
Epoch_x=[]
for epoch in range(num_epochs):
    batch=0
    running_loss = 0.0
    for xb,yb in training_generator:
        xb=xb.to(device)
        yb=yb.to(device)
        #forward feed
        y_pred = model(xb.requires_grad_())
        i=0
        for a in y_pred:
            i+=1
        #print(i)
        loss= loss_function(y_pred, yb)

        #backward propagation: calculate gradients
        loss.backward()

        #update the weights
        optimizer.step()

        #clear out the gradients from the last step loss.backward()
        optimizer.zero_grad()
        batch+=1
        running_loss += loss.item()
        Epoch_x.append(float(epoch+(batch/len(training_generator))))
        Loss_y.append(float(loss.item()))
        print('epoch {}, batch {}, loss {}'.format(epoch, batch,loss.item()))
    epoch_loss = running_loss / len(training_generator)
    losses.append(epoch_loss)
```

```python
plt.plot(Epoch_x,Loss_y)
```

```python
def cal_acc(a,b):
    ans=0
    for i in range(0,len(a)):
        c=a[i].cpu().detach().numpy()
        d=b[i].cpu().detach().numpy()
        if (d>0.6 and c==1) or (d<0.6 and c==0) :
            ans+=1
    return (ans/len(a))
```

```python
x_test=x_test.to_numpy()
y_test=y_test.to_numpy()
x_test = torch.from_numpy(x_test.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
params = {'batch_size': 64,
          'shuffle': False,
          'num_workers': 6}
test_set=torch.utils.data.TensorDataset(x_train,y_train)
test_generator = torch.utils.data.DataLoader(training_set, **params)
```

```python
a=[]
b=[]
i=1
for xb,yb in test_generator:
    xb=xb.to(device)
    yb=yb.to(device)
    #forward feed
    y_pred = model(xb)
    a.append(i)
    i+=1
    b.append(cal_acc(yb,y_pred))
print(b)
```
